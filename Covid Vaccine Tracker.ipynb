Import Libaries 

```python
import numpy as np
import pandas as pd
import plotly as py
import plotly.express as px
import plotly.graph_objs as go
from plotly.subplots import make_subplots
from plotly.offline import download_plotlyjs
#init_notebook_mode, plot, inplot
#init_notebook_mode(connected=True)
```

Import Data

```python
df= pd.read_csv("coviddata.csv")
df= df.rename (columns={'location':'Country'})

df.head()
```

```python
dfCountries= df.groupby(['Country', 'date']).sum().reset_index().sort_values('date', ascending= False)
dfCountries= dfCountries.drop_duplicates(subset=['Country'])
dfCountries= dfCountries[dfCountries['people_vaccinated']>0]
dfCountries
```

```python
fig = go.Figure(data=go.Choropleth(
    locations = dfCountries['Country'],
    locationmode = 'country names',
    z = dfCountries['people_vaccinated'],
    colorscale = 'Reds',
    marker_line_color = 'black',
    marker_line_width = 0.5,
))
```

```python
fig.update_layout(
    title_text = 'Confirmed People Vaccinated as of April 29, 2021',
    title_x = 0.5,
    geo=dict(
        showframe = False,
        showcoastlines = False,
        projection_type = 'equirectangular'
    )
)
```

```python
dfCountrydate = df[df['people_vaccinated']>0]
dfCountrydate = dfCountrydate.groupby(['date','Country']).sum().reset_index()
dfCountrydate
```

```python
fig = px.choropleth(dfCountrydate, 
                    locations="Country", 
                    locationmode = "country names",
                    color="people_vaccinated", 
                    hover_name="Country", 
                    animation_frame="date"
                   )
```

```python
fig.update_layout(
    title_text = 'Number of Vaccinated People Around the World',
    title_x = 0.5,
    geo=dict(
        showframe = False,
        showcoastlines = False,
    ))
    
fig.show()
```
